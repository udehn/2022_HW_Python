```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm, expon, poisson, binom
```

# Равномерное распределение

```python
from scipy.stats import uniform

x = np.linspace(-1, 2, 100)

plt.plot(x, uniform.cdf(x))
plt.title('График функции равномерного распределения')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid()
```

```python
plt.plot(x, uniform.pdf(x))
plt.xlabel('x')
plt.ylabel('rho(x)')
plt.title('График плотности вероятности равномерного распределения')
plt.grid()
```

# Биномиальное распределение

```python
from scipy.stats import binom

n = 100
p = 0.618
```

```python
x = np.linspace(0, 99, 100)

plt.plot(x, binom.cdf(x, n, p))
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('График функции биномиального распределения')
plt.grid()
```

```python
x = np.linspace(0, 99, 100)

plt.scatter(x, binom.pmf(x, n, p))
plt.xlabel('x')
plt.ylabel('rho(x)')
plt.title('График плотности вероятности биномиального распределения')
plt.grid()
```
